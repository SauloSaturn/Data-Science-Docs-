A freqüência relativa de um
valor é a fração ou proporção de vezes em que o valor ocorre. Denotada por:


$$ f_i = \frac{n_i} {n} $$

Onde $n_i$ é a frequência absoluta e $n$ é a dimensão do conjunto de dados.



```python
# Calculando a dimensão do conjunto de dados (somatório).
y = [1144, 1163, 885, 14955, 1448, 1425]
Dimensao_conj = sum(y)
# Calculando a frequência relativa.
Frequencia_relativa = []
for Frequencia_absoluta in y:
  Frequencia_relativa.append(Frequencia_absoluta/Dimensao_conj)
print(Frequencia_relativa)
```

### <font color = pink > 1.1.2 Gráfico Histograma

***Histograma***:
Um histograma é uma representação gráfica de uma distribuição de frequência.
Usado principalmente para dados contínuos, o eixo horizontal é dividido em intervalos chamados de "bins"(também conhecido como dados agrupados), e a altura de cada barra representa a frequência ou densidade da ocorrência dos dados dentro do bin.
É útil para visualizar a forma geral e a dispersão dos dados.

```python
import matplotlib.pyplot as plt
import numpy as np

#Data
x = list(range(2007, 2013))
y = [1144, 1163, 885, 14955, 1448, 1425]

# Calculando soma acumulativa de intervalos de dois anos.
cumulative_sum = []

for i in range(0, len(y), 2):
    interval_sum = sum(y[i:i+2]) # Sub-lista de intervalos em pares.
    cumulative_sum.append(interval_sum) # Somando inervalos e adicionando em outra lista.


# Potando o histograma com anotações textuais em razão das somas acumulativas.
plt.bar(x[::2], cumulative_sum, edgecolor='deeppink', color='thistle', width=1.8)  # Use x[::2] to get even years (FATIAMENTO DE DOIS EM DOIS)
plt.gca().set_facecolor('whitesmoke')

# Adicionando legendas
plt.xlabel('Ano')
plt.ylabel('Soma acumulativa')
plt.title('Histograma acumulativo de notificações de dois em dois anos')

# Adicionando os valres de cada barra do histograma.
for i, value in enumerate(cumulative_sum):
    plt.text(x[::2][i], value + 50, str(value), ha='center', va='bottom') #value+50 serve para que o texto não seja sobreposto à linha  da barra.

# Plot
plt.show()
```

## <font color = pink > 2. Média Aritimética

* Média Aritmética Simples

A média aritimética é a soma de todos os elementos de um conjunto em razão da quantidade de elementos do mesmo. Existem dois tipos de média para se calcular em um conjunto de dados, são elas: **Média Populacional** ($\mu$) e **Média Amostral**( $\bar{x}$), representadas pela equação abaixo.

$$ \mu = \sum_{i = 1}^{N}\frac{x_{i}} { N}
$$

onde $\bar{x} = \mu$, dependendo do espaço amostral usado basta trocar a variável.


```python
import pandas as pd
# Data
Ano = [2007, 2008, 2009, 2010, 2011, 2012]
Notificações = [1144, 1163, 885, 14955, 1448, 1425]

# Soma de todos os elementos.
Soma_ele = sum(Notificações)

# Determinando a população.
Pop = len(Ano)

# Cálculo da média.

Med_Pop = Soma_ele / Pop

print(Med_Pop)
```

* Média Agrupada

A média agrupada é o somatório dos produtos das médias dos intervalos com suas respectivas frequências, onde o somatório é dividido pelas frequência total.

$$ \mu = \frac{1}{N}\sum_{i = 1}^{n}  m_{i} \cdot f_{i} $$

Calculando a média agrupada do nosso data set.

**Parte I**: Determinando os pontos médios $m_i$ e as frequências $f_i$.

```python
import pandas as pd

# Data.
Ano = [2007, 2008, 2009, 2010, 2011, 2012]
Notificações = [1144, 1163, 885, 14955, 1448, 1425]
Ano_agr1= ['2007 - 2008', '2009 - 2010', '2011 -2012']

# Criando o mi (ponto médio dos intervalos)
Ano_agr2 = []
Ano_mi = []
for i in range (0, len(Ano), 2):
  interval_mi = sum(Ano[i:i+2])
  Ano_agr2.append(interval_mi)
for i in Ano_agr2:
    i = (int(i))/2
    Ano_mi.append(i)

#print(Ano_agr2)
#print(Ano_mi)

# Criando o fi (fi é a soma das not dois a dois)
Notificações = [1144 + 1163, 885 + 14955, 1448 + 1425]

#print(Notificações)

#Plot Data Frame mi e fi.
d = {'Ano': Ano_agr1, 'Notificações':Notificações} # Criando um dic para adicionar novas colunas (listas).
df = pd.DataFrame(d)
df['Ano_mi'] = Ano_mi # Adicionando no df a coluna ano mi.
df['Not_fi'] = Notificações # Adicionando no df a coluna not fi.
df
```

**Parte II**: Determinando  o produto entre os pontos médios e as frequências $m_i \cdot f_i$ e a média agrupada $\mu$.

```python
import pandas as pd
# Operando mi * fi e adicionando no df.
df.loc[:,'mi * fi'] = (df.loc[:,'Ano_mi']).mul(df.loc[:,'Not_fi'])# Selecionando e operando uma coluna.
df
```

```python
# Calculando a média agrupada.
soma_not = sum(df.loc[:,'Notificações'])
soma_mifi = sum(df.loc[:,'mi * fi'])
media_agr = soma_mifi / soma_not
print("Média de dados agrupados = ",media_agr)
```

## <font color = pink > 2. Variância e Desvio Padrão
* Variância e Desvio Padrão Simples

A variância, $\sigma^2$, é a média dos quadrados dos desvios entre os valores observados e a média. A variância é cálculada por:

$$ \sigma^2 = \frac{1}{n}\sum_{i = 1}^{n}  {(x_{i}} - \bar {x})^2 $$


O Desvio padrão é utilizado para corrigir a disparidade da variância, e portanto seu cálculo basta fazer a raíz quadrada da própria variância.

$$ \sigma = \sqrt{\frac{1}{n}\sum_{i = 1}^{n}  ({x_{i}} - \bar {x})^2} $$


```python
import pandas as pd
# Data.
Ano = [2007, 2008, 2009, 2010, 2011, 2012]
Notificações = [1144, 1163, 885, 14955, 1448, 1425]

# Definindo o data frame.
df = pd.DataFrame(Notificações, columns = ['Notificações'])
df

# Chamando o comando mean(), para calcular a média.
Media_df = df.loc[:, 'Notificações'].mean()

# Calculando o quadrado.
Quad = []
for i in Notificações:
  Quad.append((i**2 - Media_df**2))
print(Quad)

# Calculando a variância.
Var_df = (sum(Quad))/len(Notificações)
print("A viarância é:", Var_df)

# Calculando o desvio padrão.
Desv_df = (Var_df)**0.5
print("O desvio padrão é:", Desv_df)
```

## <font color = pink > 2.1.Permutação Simples (Fatorial)

**Definição**

 O número de modos de ordenar $n$ objetos distintos é:


 $$n(n - 1) ... 3\cdot 2\cdot 1 = n!$$

 ou simplismente um produtório:

 $$ \prod_{i = 1}^{n}i = n!$$
  
  Também podemos dizer que um fatorial é um caso particular de um *arranjo*, é dizer, uma *permutação simples*.

  $$ A_{p}^{p} = n!$$




```python
n = int(input("Digite um fatorial a ser calculado: "))
contador = 0 # contador precisa ser começar do 0, pois é o elemento nulo da adição.
produto = 1 # produto precisa ser começar do 1, pois é o elemento nulo da multiplicação.
while contador < n:
  contador = contador + 1 # O valor altera a cada loop e armazena em contador.
  produto = produto * contador # O valor altera a cada loop e armazena em produto.
print("Fatorial é: ", produto) # Printando o produto.
```

## 2.2.Arranjo Simples
Dado um conjunto com $n$ elementos distintos: $\{ a_1, a_2, a_3, ... , a_n \}$, chama-se arranjo simples desses $n$ elementos, tomados $p$ a $p$ (com $p ≤ n$),todo agrupamento ordenado de $p$  elementos distintos escolhidos dentre os $n$ elementos, onde a **ordem** dos elementos **importa**. Segue-se que:

$$ A _ {p}^{n} = \frac{n!} {(n-p)!} $$

 Lê-se: Arranjo de $n$ elementos tomados $p$ a $p$.

```python
n = int(input('Digite a quantidade n de elementos: '))
p = int(input('Digite a quantidade p de elementos: '))
contador = n
produto = 1
while contador > (n - p): # Restrição do laço até a diferença do denominador.
  produto = produto * contador # Multiplicando até a diferença.
  contador = contador - 1
print("O arranjo simples é: ",produto)
```

## 2.3.Combinação Simples
Dado um conjunto com $n$ elementos distintos: $\{ a_1, a_2, a_3, ... , a_n \}$, chama-se combinação simples desses $n$ elementos, tomados $p$ a $p$ (com $p ≤ n$), todo subconjunto de $p$ elementos distintos escolhidos dentre os $n$ elementos. Onde a **ordem** dos elementos **não importa**. Segue-se que:

$$ \begin{pmatrix} n \\ p\end{pmatrix} = \frac{n!} {p! (n-p)!}
$$

ou simplesmente:

$$ C_{n}^{p} = \frac{n!} {p! (n-p)!}$$

```python
n = int(input('Digite a quantidade n de elementos: '))
p = int(input('Digite a quantidade p de elementos: '))

contador_1 = n # Contador para o numerador.
contador_2 = p # Contador para o denominador.
produto_1 = 1
produto_2 = 1

while contador_1 > (n-p): # While para o numerador.
  produto_1 = produto_1 * contador_1
  contador_1 = contador_1 - 1


while contador_2 > 1: # While para o denominador.
     produto_2 = produto_2 * contador_2
     contador_2 = contador_2 - 1

print("A combinação simples é: ",int(produto_1/produto_2))
```

## 2.4.Permutação Circular
A permutação circular é um caso de permutação em que os $n$ elementos estão dispostos em um ciclo. Na permutação circular, a **ordem** em que um **elemento segue o outro é importante**. Segue-se que:

$$(PC)_{n} = (n - 1)!$$


```python
n = int(input('Digite a quantidade n de elementos: '))

contador = n - 1  # Adicionando o n - 1 para o contador.
produto = 1

while contador > 1: # While idêntico ao do fatorial (permutação simples).
  produto = produto * contador
  contador = contador - 1
print("A permutação circular é: ",produto)
```

## 2.5.Permutação por Repetição
Considere $n$ elementos, nem todos distintos, de um conjunto $\{ a_1, a_2, a_3, ... , a_n \}$. O
número de permutações formado com todos os $n$ elementos desse conjunto é dado por:

$$P _ {n}^{n_1,n_2,n_3,...} = \frac{n!} {n_1!n_2! ... n_k!}$$

Onde $n_1,n_2,n_3,...$ indicam o número de repetições de cada elemento do conjunto de
$n$ elementos.

Para a modelagem da permutação por repetição usaremos uma função suporte chamada **fato(n)**, a qual criaremos do zero, seguindo o código similar ao da permutação simples.

```python
def fato(n):

  contador = n #o contador é necessário para fazer a regressão fatorial.
  produto = 1 #a variável produto deve ser 1.

  while contador > 1:
     produto *= contador
     contador = contador - 1
  return(produto)
```

O código referente à permutação por repetição sgue-se abaixo.

```python
permutação = int(input('Insira o número de letras: '))

booleano = True # Criando um boolean para validar o laço while.
repeticao = [] # Criando uma lista vazia para armazenar as repetições.
lista_negação = ['não', 'NÃO', 'Não', 'nao', 'Nao', 'NAO', 'no', 'NO']# Criando uma lista para sair do laço while.

while booleano == True:  # Criando um laço while para inputs múltiplos.
  pergunta_1 = int(input('Insira o número de repetições da letra '))
  repeticao.append(pergunta_1)
  pergunta_2 = input('Continuar? Sim ou Não?: ')
  if pergunta_2 in lista_negação:
    booleano = False
print(repeticao)

contador_i = 1
produto_i = 1
for i in repeticao: # Criando um for para percorrer a lisa das repetições usando a função fato().
  i = fato(i)
  produto_i = produto_i * i

permutacao_repeticao = fato(permutação)/produto_i # Calculando a permutação por repetição com os resultados acima.
print(int(permutacao_repeticao))
```
